## 연도별 중요 Key word 확인 
(method: TF-IDF)

```python
# TF-IDF 벡터화
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer()
```

```python
# 연도별 후보자별 TF-IDF 계산 함수 정의
def calculate_tfidf_by_year_and_candidate(df):
    results = {}
    idf_results = {}
    years = df['Year']
    for year in years:
        candidates_in_year = df[df['Year'] == year]['Name']
        for candidate in candidates_in_year:
            subset_df = df[(df['Year'] == year) & (df['Name'] == candidate)]
            vectorizer = TfidfVectorizer()
            tfidf_matrix = vectorizer.fit_transform(subset_df['speech_processed2'])
            feature_names = vectorizer.get_feature_names_out()
            dense = tfidf_matrix.todense()
            df_tfidf = pd.DataFrame(dense, columns=feature_names)
            # 평균 TF-IDF 계산 후 내림차순 정렬
            avg_tfidf = df_tfidf.mean().sort_values(ascending=False)
            results[f"Year {year}, Candidate {candidate}"] = avg_tfidf
            # IDF 값 저장
            idf_values = vectorizer.idf_
            idf_results[f"Year {year}, Candidate {candidate}"] = dict(zip(feature_names, idf_values))
    return results, idf_results
```

```python
# 연도별 후보자별 TF-IDF 및 IDF 계산
yearly_candidate_tfidf, yearly_candidate_idf = calculate_tfidf_by_year_and_candidate(df)
```

```python
# 연도별 후보자별 TF-IDF 계산 함수
def calculate_tfidf_by_group(df, group_col):
    results = {}
    groups = df.groupby(group_col)
    for group_name, group_data in groups:
        vectorizer = TfidfVectorizer()
        tfidf_matrix = vectorizer.fit_transform(group_data['speech_processed2'])
        feature_names = vectorizer.get_feature_names_out()
        dense = tfidf_matrix.todense()
        df_tfidf = pd.DataFrame(dense, columns=feature_names)
        # 평균 TF-IDF 계산 후 내림차순 정렬
        avg_tfidf = df_tfidf.mean().sort_values(ascending=False)
        results[group_name] = avg_tfidf
    return results
```

```python
# 연도별 TF-IDF 계산
yearly_tfidf = calculate_tfidf_by_group(df, 'Year')

# 후보자별 TF-IDF 계산
candidate_tfidf = calculate_tfidf_by_group(df, 'Name')

# 결과 출력
print("연도별 TF-IDF:")
for year, tfidf in yearly_tfidf.items():
    print(f"Year {year}:")
    print(tfidf.head(10))  # 상위 10개 단어 출력

print("\n후보자별 TF-IDF:")
for candidate, tfidf in candidate_tfidf.items():
    print(f"Candidate {candidate}:")
    print(tfidf.head(10))  # 상위 10개 단어 출력
```

```python
# 1.speech_processed2로 분석

def calculate_tfidf_by_party(df):
    party_tfidf = {}
    vectorizer = TfidfVectorizer()
    
    # Party별로 데이터를 분리
    for party in df['Party'].unique():
        # 각 Party에 대해 speech_processed2를 합치기
        party_speeches = df[df['Party'] == party]['speech_processed2'].str.cat(sep=' ')
        
        # TF-IDF 계산
        tfidf_matrix = vectorizer.fit_transform([party_speeches])
        feature_names = vectorizer.get_feature_names_out()
        dense = tfidf_matrix.todense()
        df_tfidf = pd.DataFrame(dense, columns=feature_names)
        
        party_tfidf[party] = df_tfidf

    return party_tfidf, vectorizer
```

```python
def print_top_keywords(party_tfidf, top_n=10):
    for party, df_tfidf in party_tfidf.items():
        print(f"Party: {party}")
        top_words_idx = df_tfidf.mean().sort_values(ascending=False).head(top_n).index
        top_words_values = df_tfidf.mean().sort_values(ascending=False).head(top_n).values
        top_words_df = pd.DataFrame({'단어': top_words_idx, 'TF-IDF 값': top_words_values})
        print(top_words_df)
        print()
```

```python
import numpy as np


# TF-IDF 분석을 위한 함수 정의
def tfidf_analysis(df, name):
    # 결합된 연설문 텍스트 추출
    speeches = df['speech_processed2'].dropna().tolist()
    
    # TF-IDF 벡터라이저 초기화
    vectorizer = TfidfVectorizer(stop_words='english', max_features=100)  # 상위 100개 단어만 추출
    X = vectorizer.fit_transform(speeches)
    
    # TF-IDF 스코어 합계 계산
    tfidf_scores = np.sum(X.toarray(), axis=0)
    # 단어와 해당 스코어 매핑
    feature_names = vectorizer.get_feature_names_out()
    tfidf_dict = dict(zip(feature_names, tfidf_scores))
    
    # TF-IDF 스코어에 따라 정렬된 상위 10개 단어 출력
    sorted_tfidf = sorted(tfidf_dict.items(), key=lambda item: item[1], reverse=True)[:20]
    print(f"{name}의 상위 30개 중요 단어:")
    for word, score in sorted_tfidf:
        print(f"{word}: {score:.4f}")
    
    return sorted_tfidf
```

```python
# Joseph R. Biden 연설문 중요도 분석
print("Joseph R. Biden 연설문 중요도 분석:")
biden_tfidf = tfidf_analysis(biden_df, 'Joseph R. Biden')
```

```python
# Donald J. Trump 연설문 중요도 분석
print("\nDonald J. Trump 연설문 중요도 분석:")
trump_tfidf = tfidf_analysis(trump_df, 'Donald J. Trump')
```

```python
# 시각화를 위한 데이터 준비
biden_words, biden_scores = zip(*biden_tfidf)
trump_words, trump_scores = zip(*trump_tfidf)

# 그래프 그리기
fig, ax = plt.subplots(figsize=(12, 8))

# 바이든의 중요 단어들을 파란색으로 표시
ax.barh(biden_words, biden_scores, color='blue', alpha=0.6, label='Joseph R. Biden')

# 트럼프의 중요 단어들을 빨간색으로 표시
ax.barh(trump_words, trump_scores, color='red', alpha=0.6, label='Donald J. Trump')

# 그래프 스타일 설정
ax.set_xlabel('TF-IDF Score', fontsize=12)
ax.set_ylabel('단어', fontsize=12)
ax.set_title('Joseph R. Biden vs. Donald J. Trump 연설문 중요 단어', fontsize=14)
ax.legend()

# 그래프 출력
plt.gca().invert_yaxis()  # 단어를 내림차순으로 표시하기 위해 y축 역순으로 설정
plt.tight_layout()
plt.show()
```

```python
# 연설문 데이터를 전처리하여 불필요한 단어 제거
def preprocess_text(texts):
    processed_texts = []
    for text in texts:
        text = text.lower()
        processed_texts.append(text)
    return processed_texts

biden_speeches = preprocess_text(biden_speeches)
trump_speeches = preprocess_text(trump_speeches)

# TF-IDF 벡터라이저 초기화
vectorizer_biden = TfidfVectorizer(stop_words='english', max_df=0.85, min_df=1)
X_biden = vectorizer_biden.fit_transform(biden_speeches)

vectorizer_trump = TfidfVectorizer(stop_words='english', max_df=0.85, min_df=1)
X_trump = vectorizer_trump.fit_transform(trump_speeches)
```

```python
# LDA 모델 초기화 (주제 개수 설정)
n_topics = 5
lda_biden = LatentDirichletAllocation(n_components=n_topics, max_iter=10, learning_method='online', random_state=42)
lda_trump = LatentDirichletAllocation(n_components=n_topics, max_iter=10, learning_method='online', random_state=42)

# LDA 모델 fitting
lda_biden.fit(X_biden)
lda_trump.fit(X_trump)

# LDA 시각화
pyLDAvis.enable_notebook()

# Feature names 가져오기
feature_names_biden = vectorizer_biden.get_feature_names_out()
feature_names_trump = vectorizer_trump.get_feature_names_out()

# pyLDAvis 데이터 준비 함수
def prepare_lda_vis(lda_model, dtm, feature_names):
    topic_term_dists = lda_model.components_ / lda_model.components_.sum(axis=1)[:, None]
    doc_topic_dists = lda_model.transform(dtm)
    doc_lengths = dtm.sum(axis=1).getA1()
    term_frequency = dtm.sum(axis=0).getA1()

    return pyLDAvis.prepare(
        topic_term_dists,
        doc_topic_dists,
        doc_lengths,
        feature_names,
        term_frequency
    )

# 바이든의 LDA 시각화 준비
data_biden = prepare_lda_vis(lda_biden, X_biden, feature_names_biden)
pyLDAvis.save_html(data_biden, 'biden_lda.html')
pyLDAvis.display(data_biden)

# 트럼프의 LDA 시각화 준비
data_trump = prepare_lda_vis(lda_trump, X_trump, feature_names_trump)
pyLDAvis.save_html(data_trump, 'trump_lda.html')
pyLDAvis.display(data_trump)
```
