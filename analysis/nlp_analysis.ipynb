```python
from bertopic import BERTopic
import ast


from bertopic.representation import KeyBERTInspired

from sklearn.feature_extraction.text import CountVectorizer



def do_bertopic(df):
  vectorizer_model = CountVectorizer(stop_words="english")
  representation_model = KeyBERTInspired()
  topic_model = BERTopic(n_gram_range=(1,2), representation_model=representation_model, vectorizer_model=vectorizer_model)

  topics, probs = topic_model.fit_transform(df["speech_whole"])

  return topic_model, topics, probs
```

```python
import re
def filter_externals(string):
    # externals_list = ['[applause]', '(applause)']
    # for external in externals_list:
    #     string = string.replace(external, '')

    # [ (  ) ]
    string = re.sub('\s*\[[^]]*\]', '', string)
    string = re.sub('\s*\([^)]*\)', '', string)
    string = re.sub('\s*\([^]]*\]', '', string)
    string = re.sub('\s*\[[^)]*\)', '', string)
    return string


vectorizer_model = CountVectorizer(stop_words="english")
representation_model = KeyBERTInspired()
topic_model = BERTopic(n_gram_range=(1,2), representation_model=representation_model, vectorizer_model=vectorizer_model)

topics, probs = topic_model.fit_transform(df_all["speech_whole"].apply(filter_externals))
```

## KeyBERT

```python
def do_keybert(df):
  kw_model = KeyBERT()
  df["keywords"] = df["speech_whole"].progress_apply(lambda x: kw_model.extract_keywords(x, keyphrase_ngram_range=(1, 2), stop_words='english'))
  return kw_model, df
```

## Vader

```python

# nltk.download('vader_lexicon')

def do_vader(df):
    sid = SentimentIntensityAnalyzer()
    df["sentiment"] = df["speech_whole"].progress_apply(lambda x: sid.polarity_scores(x))
    return sid, df
```
