```python
import pandas as pd
```

```python
# 파일 경로
approval_path = '/content/US_Presidential_Approval_1913_2024.csv'
elections_path = '/content/us_presidents_elections_split.csv'
gdp_path = '/content/US_GDP_Data_1913-2023.csv'
cpi_path = '/content/US_CPI_Data_1913-2023.csv'
employment_path = '/content/Employment_Status.csv'
```

```python
# 파일 업로
approval_df = pd.read_csv(approval_path)
elections_df = pd.read_csv(elections_path)
gdp_df = pd.read_csv(gdp_path)
cpi_df = pd.read_csv(cpi_path)
employment_df = pd.read_csv(employment_path)
```

```python
# 열 이름 변경
elections_df.rename(columns={'Election_Year': 'Year'}, inplace=True)
```

```python
# 데이터 meaged
merged_df = approval_df.merge(elections_df, on='Year', how='outer')
merged_df = merged_df.merge(gdp_df, on='Year', how='outer')
merged_df = merged_df.merge(cpi_df, on='Year', how='outer')
final_merged_df = merged_df.merge(employment_df, on='Year', how='outer')
```

```python
# 필요한 행 추출하여 병합
final_merged_df['Name'] = final_merged_df['Name'].fillna(method='ffill')
final_merged_df['Elections_Won'] = final_merged_df['Elections_Won'].fillna(method='ffill')
```

```python
final_merged_df.head()
```

```python
# 필요없는 열 지우기
final_merged_df = final_merged_df.iloc[:112]
```

```python
# 열 이름 변경
final_merged_df.rename(columns={'Percent of labor force': 'Employment status'}, inplace=True)
```

```python
# CPI, GDP, Employment status 결측치 채우기
final_merged_df.loc[final_merged_df['Year'] == 2024, 'CPI'] = 312.1445
final_merged_df.loc[final_merged_df['Year'] == 2024, 'GDP'] = 22936322
final_merged_df.loc[final_merged_df['Year'] == 2023, 'Employment status'] = 3.7
final_merged_df.loc[final_merged_df['Year'] == 2024, 'Employment status'] = 4.0
```
