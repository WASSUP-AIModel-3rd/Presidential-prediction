```python
import pandas as pd

df_2024 = pd.read_csv('2024_combined_ver2.csv')
df_2020 = pd.read_csv('text_combined.csv')
```

```python
df_2020 = df_2020[df_2020['Year'] == 2020]
```

```python
# name 별로 데이터프레임 분리
biden_df_2024 = df_2024[df_2024['name'] == 'Joseph R. Biden']
trump_df_2024 = df_2024[df_2024['name'] == 'Donald J. Trump']
biden_df_2020 = df_2020[df_2020['name'] == 'Joseph R. Biden']
trump_df_2020 = df_2020[df_2020['name'] == 'Donald J. Trump']
```

```python
# 2024년과 2002년의 데이터프레임 결합
biden_df = pd.concat([biden_df_2024, biden_df_2020])
trump_df = pd.concat([trump_df_2024, trump_df_2020])
```

```python
biden_df
```

```python
with open("idflist","rb") as fp:
   idf_list = pickle.load(fp)[-100:]

def filter_lowidf(token_list):
    token_list = ast.literal_eval(token_list)
    filtered_list = list()

    for token in token_list:
        if token not in filter_words:
            filtered_list.append(token)
    return filtered_list

biden_df["speech_processed3"] = biden_df["speech_processed"].apply(filter_lowidf) 
trump_df["speech_processed3"] = trump_df["speech_processed"].apply(filter_lowidf) 
```

```python
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer

df['speech_processed3'] = df['speech_processed3'].astype(str)

# TF-IDF 분석을 위한 함수 정의
def tfidf_analysis(df, name):
    # 결합된 연설문 텍스트 추출
    speeches = df['speech_processed3'].dropna().tolist()
    
    # 각 문서가 문자열인지 확인 (추가적인 안전 장치)
    if not all(isinstance(text, str) for text in speeches):
        raise ValueError("All entries in 'speech_processed3' must be strings.")
    
    # TF-IDF 벡터라이저 초기화
    vectorizer = TfidfVectorizer(stop_words='english', max_features=100)  # 상위 100개 단어만 추출
    X = vectorizer.fit_transform(speeches)
    
    # TF-IDF 스코어 합계 계산
    tfidf_scores = np.sum(X.toarray(), axis=0)
    # 단어와 해당 스코어 매핑
    feature_names = vectorizer.get_feature_names_out()
    tfidf_dict = dict(zip(feature_names, tfidf_scores))
    
    # TF-IDF 스코어에 따라 정렬된 상위 30개 단어 출력
    sorted_tfidf = sorted(tfidf_dict.items(), key=lambda item: item[1], reverse=True)[:30]
    print(f"{name}의 상위 30개 중요 단어:")
    for word, score in sorted_tfidf:
        print(f"{word}: {score:.4f}")
    
    return sorted_tfidf
```

```python
print("Joseph R. Biden 연설문 중요도 분석:")
biden_tfidf = tfidf_analysis(biden_df, 'Joseph R. Biden')
```

```python
# Donald J. Trump 연설문 중요도 분석
print("\nDonald J. Trump 연설문 중요도 분석:")
trump_tfidf = tfidf_analysis(trump_df, 'Donald J. Trump')
```

```python
import matplotlib.pyplot as plt

# 시각화를 위한 데이터 준비
biden_words, biden_scores = zip(*biden_tfidf)
trump_words, trump_scores = zip(*trump_tfidf)

# 그래프 그리기
fig, ax = plt.subplots(figsize=(12, 8))

# 바이든의 중요 단어들을 파란색으로 표시
ax.barh(biden_words, biden_scores, color='blue', alpha=0.6, label='Joseph R. Biden')

# 트럼프의 중요 단어들을 빨간색으로 표시
ax.barh(trump_words, trump_scores, color='red', alpha=0.6, label='Donald J. Trump')

# 그래프 스타일 설정
ax.set_xlabel('TF-IDF Score', fontsize=12)
ax.set_ylabel('단어', fontsize=12)
ax.set_title('Joseph R. Biden vs. Donald J. Trump 연설문 중요 단어', fontsize=14)
ax.legend()

# 그래프 출력
plt.gca().invert_yaxis()  # 단어를 내림차순으로 표시하기 위해 y축 역순으로 설정
plt.tight_layout()
plt.show()
```
