```python
def crawl_campaign_speech_by_year(year, names, election, url, mode=["archive","archive"]):
    for i, m in enumerate(mode):
        if m:
            if m == "archive":
                crawl = nomination_crawler.crawl_campaign_speech
            else:
                crawl = nomination_crawler.crawl_campaign_speech_from_search

            df = crawl(url=url[i],year=year, save=True, save_path=f"./data/{year}_{names[i]}_speech.csv")
            df["elected"] = election[i]
            df.to_csv(f"./data/{year}_{names[i]}_speech_edited.csv")
```

## 후보연설과 매칭

```python
df2 = pd.read_csv('./Nomination_speech.csv', index_col=0)
```

```python
df2["elected"] = 0
```

```python
for idx, data in enumerate(df2.values):
    name = data[0]
    year = int(data[1].split(' ')[-1])

    if year >= 2000:
        start = 41
    elif year >= 1900:
        start = 16
    else:
        start = 0
    for result in elected_list[start:]:
        if result[0]-year <= 2:
            if str.lower(result[1]) in str.lower(name):
                df2["elected"].iloc[idx] = 1
                break
```

```python
df2.to_csv("Nomination_speech_edited.csv")
```
